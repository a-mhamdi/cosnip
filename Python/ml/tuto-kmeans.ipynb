```python
import pandas as pd
```

```python
# Click `copy csv` before running this cell
df = pd.read_clipboard(sep=',')
```

```python
df.columns
```

```python
df.head(3)
```

```python
n_clusters = len(set(df.z))
```

```python
n_clusters
```

```python
import seaborn as sns
```

```python
sns.set_palette('hls')
```

```python
sns.color_palette()
```

```python
sns.scatterplot(df, x='x', y='y', hue='z');
```

```python
import sklearn
```

```python
from sklearn.cluster import KMeans
```

```python
clf = KMeans(n_clusters, init='k-means++')
```

```python
data = df.drop(columns='z')
```

```python
clf.fit(data)
```

```python
out = clf.predict(data)
```

```python
import matplotlib.pyplot as plt
```

```python
_, ax = plt.subplots(1, 2)
sns.scatterplot(df, x='x', y='y', hue='z', ax=ax[0]) # Ground Truth
sns.scatterplot(data, x='x', y='y', hue=out, ax=ax[1]) # Estimated Clustering
```
